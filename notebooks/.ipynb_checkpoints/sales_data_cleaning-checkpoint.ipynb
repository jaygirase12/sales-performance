```python
import pandas as pd
import numpy as np
df = pd.read_csv(r"C:\Data Analyst Projects\sales-performance\data\Superstore.csv", encoding="latin1")
df.head()
```

```python
df.isnull().sum()
```

```python
# Drop duplicate rows if any
df = df.drop_duplicates()
```

```python
# Convert dates to datetime
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Ship Date'] = pd.to_datetime(df['Ship Date'])

# Ensure Postal Code is string (not int)
df['Postal Code'] = df['Postal Code'].astype(str)
```

```python
df['Profit Margin'] = df['Profit'] / df['Sales']
df['Order Month'] = df['Order Date'].dt.month
df['Order Year'] = df['Order Date'].dt.year
```

```python
# profit margin
df['Profit Margin %'] = (df['Profit'] / df['Sales']) * 100
# sales per quantity(unit price)
df['Sales per Quantity'] = df['Sales'] / df['Quantity']
#shipping delays(days)
df['Shipping Delay'] = (df['Ship Date'] - df['Order Date']).dt.days
```

```python
df.to_csv("../data/superstore_cleaned.csv", index=False)
```
